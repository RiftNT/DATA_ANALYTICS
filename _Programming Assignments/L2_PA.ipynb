```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('diabetes.csv')

# replace string to num values
df.replace({'Yes': 1, 'No': 0}, inplace=True)
df['Gender'].replace({'Male': 1, 'Female': 0}, inplace=True)
df['class'].replace({'Positive': 1, 'Negative': 0}, inplace=True)

df.head()
```

```python
# histogram plot
plt.figure(figsize=(12, 10))
df.hist(figsize=(12, 10))
plt.tight_layout()
plt.show()
```

```python
# correlation matrix
plt.figure(figsize=(10, 8))
plt.matshow(df.corr(), cmap='coolwarm')
plt.title('Correlation Matrix')
plt.xticks(range(len(df.columns)), df.columns, rotation='vertical')
plt.yticks(range(len(df.columns)), df.columns)
plt.colorbar()
plt.show()
```

```python
# pie chart
plt.figure(figsize=(6, 6))
df['class'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Diabetes Class Distribution')
plt.ylabel('')
plt.show()
```

```python
# stacked bar charts
gender_class = df.groupby(['sudden weight loss', 'class']).size().unstack()
gender_class.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Diabetes Class Distribution by sudden weight loss')
plt.xlabel('sudden_weight_loss')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Yes', 'No'], rotation=0)
plt.legend(title='Class', labels=['Negative', 'Positive'])
plt.show()
```
